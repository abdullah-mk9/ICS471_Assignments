```python
import numpy as np
from matplotlib import pyplot as plt

xs = np.linspace(-5, 5, 100)
ys = np.linspace(-5, 5, 100)

# plt.plot()
xs, ys = np.meshgrid(xs, ys)
xs.flatten()[:,None]
ps = np.hstack((xs.flatten()[:,None], ys.flatten()[:,None]))

w = np.array([1, 2]) /2
b = 0

psw = ps @ w + b

psw[(psw<=1) & (psw>=-1)] = 0
psw[(psw>1)] = 10
psw[(psw<-1)] = -10

plt.scatter(ps[:,0], ps[:,1], s=10, c=psw, cmap='seismic')
```
